# Melt

```python
import pandas as pd

test = pd.DataFrame({"MST": ["I", "II", "III"], "Sanjog": [5, 16, 18], "Abhas": [3, "Ab", 15], "Sameer": [4, 12, 14]})
test
```

```python
pd.melt(test)
```

```python
pd.melt(test, id_vars=["MST"], var_name="Name", value_name="Marks")
```

# Pivot

```python
exam = pd.DataFrame({"Test": ["MST-I", "MST-II", "MST-III", "END SEM"], "Sanjog": [5, 16, 18, 34], "Abhas": [3, "Ab", 15, 44], "Sameer": [4, 12, 14, 36]})
exam
```

```python
exam.pivot(index="Test", columns=[])
```

```python
exam.pivot_table(index="Test", columns="Sanjog", aggfunc="sum")
```
